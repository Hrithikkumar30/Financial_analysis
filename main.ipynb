```python
def security(y):
    if (y['Securities Account']==1) & (y['CD Account']==1):
        return 'Hols Securities and Deposite Account'
    if (y['Securities Account']==0) & (y['CD Account']==0):
        return 'Neither hols Securities nor Deposite Account'
    if (y['Securities Account']==1) & (y['CD Account']==0):
        return 'Hols only Securities and not Deposite Account'
    if (y['Securities Account']==0) & (y['CD Account']==1):
        return 'Hols only Deposite Account and not Securities'
        
```

```python
data['Account_holder_category'] = data.apply(security , axis =1)
```

```python
sns.displot(data[data['Personal Loan']==0]['Income'] , kde =True ,legend ='Income with no personal loan' )
sns.displot(data[data['Personal Loan']==1]['Income'] , kde =True , label ='Income with  personal loan')
```

```python
def plot(col1,col2,label1,label2 , title):
    sns.displot(data[data[col2]==0][col1] , kde =True ,legend ='Income with no personal loan' )
    sns.displot(data[data[col2]==1][col1] , kde =True ,legend ='Income with personal loan' )
    plt.legend()
    plt.title(title)
```

```python
plot('Income' , 'Personal Loan' , 'Income with no personal loan','Income with personal loan','Income Distribution Plot')
```

```python
plot('CCAvg','Personal Loan',
     'Credit card avg with no personal loan',
     'Credit card avg with personal loan',
     'Credit card avg Distribution')
```
